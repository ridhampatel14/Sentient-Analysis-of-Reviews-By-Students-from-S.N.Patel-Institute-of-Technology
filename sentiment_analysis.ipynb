```python
df1=pd.read_csv("/content/drive/MyDrive/preproccessed_and_translated.csv")
df1
```

```python
df1.drop(df1.columns[0],axis=1, inplace=True)
df1
```

```python
import nltk
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
result_using_NLTK=[]
for i in df1['comments']:
  temp_res=sid.polarity_scores(i)
  sor_res=sorted(temp_res.items(),key=lambda x:x[1],reverse=True)
  result_using_NLTK.append(sor_res[0][0])
```

```python
df1['analysis_using_NLTK']=result_using_NLTK
```
